Steps to create a Word2Vec

- Create Histograms
- Take most frequent words

```python
import nltk #nltk is an important library for any kind of text preprocessing
import pandas as pd #Using it to import dataset into python
```

```python
from gensim.models import Word2Vec #gensim is a library which contain Word2Vec function 
#which helps in conversion of words to vectors
from nltk.corpus import stopwords #nltk.corpus contains stopwords There are stopwords for every individual language, 
#We use this to eliminate the keywords which does not add value to the data
```

```python
train = pd.read_csv('abcnews-date-text.csv')
```

```python
train.head()
```

```python
#I am converting the column to list of texts
texts = []
texts = list(train['headline_text'])[0:10000]
```

```python
#Additionally we can do text preprocessing using re library removing numbers, converting everything to lower text
texts
```

```python
#Preparing the dataset 
#Doing word tokenize on list of words
#nltk supports word_tokenize to convert sentences into word
sentences = [nltk.word_tokenize(text) for text in texts]
```

```python
sentences
```

```python
len(sentences)
```

```python
#Removing stopwords from the sentence
for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
sentences
```

```python
model = Word2Vec(sentences,min_count = 5)
```

```python
words = model.wv.vocab
words
```

```python
#Finding word vectors
vector = model.wv['national']
vector
```

most_similar(positive=None, negative=None, topn=10, restrict_vocab=None, indexer=None)
Find the top-N most similar words. Positive words contribute positively towards the similarity, negative words negatively.

This method computes cosine similarity between a simple mean of the projection weight vectors of the given words and the vectors for each word in the model. The method corresponds to the word-analogy and distance scripts in the original word2vec implementation

```python
#Most similar words

similar = model.wv.most_similar('national')
similar
```
