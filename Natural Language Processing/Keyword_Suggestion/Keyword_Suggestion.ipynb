Steps to create a Word2Vec
- Tokenization of Sentences
- Create Histograms
- Take most frequent words
- Create a matrix with all the unique words. It also represent the occurence relation between the words

```python
import nltk
import pandas as pd
```

```python
from gensim.models import Word2Vec
from nltk.corpus import stopwords
```

```python
train = pd.read_csv('abcnews-date-text.csv')
```

```python
train.head()
```

```python
texts = []
texts = list(train['headline_text'])[0:500]
```

```python
texts
```

```python
texts
```

```python
#Preparing the dataset 
#Doing word tokenize on list of words
sentences = [nltk.word_tokenize(text) for text in texts]
```

```python
sentences
```

```python
len(sentences)
```

```python
#Removing stopwords from the sentence
for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
sentences
```

```python
model = Word2Vec(sentences,min_count = 1)
```

```python
words = model.wv.vocab
words
```

```python
#Finding word vectors
vector = model.wv['national']
vector
```

```python
#Most similar words
similar = model.wv.most_similar('national')
similar
```
